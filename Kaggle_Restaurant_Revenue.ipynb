# No missing values :)

```python
# Bringing uniformity in date format
data['Open Date'] = pd.to_datetime(data['Open Date'])
data
```

```python
# splitting date column
#data['Day'] = data['Open Date'].dt.day
data['Month'] = data['Open Date'].dt.month
data['Year'] = data['Open Date'].dt.year
data = data.drop("Open Date", axis=1)
data.head()
```

# Label Encoding ? Let's do it.

```python
data.columns
```

```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
label = LabelEncoder()

df = data.copy()

for i in ['City', 'Type', 'City Group']:
    df[i+'_lab'] = label.fit_transform(df[i])

df.head()
```

```python
df = df.drop(['Id', 'City', 'Type', 'City Group'], axis=1)
```

```python
df
```

# Standardising Values with MinMaxScaler

```python
from sklearn.preprocessing import MinMaxScaler
import re
min_max = MinMaxScaler()

feature_scale = [feature for feature in df.columns if re.match("P",feature)]
feature_scale
not_feature_scaled = [feature for feature in df.columns if feature not in feature_scale]
not_feature_scaled

df_new = pd.concat([df[not_feature_scaled].reset_index(drop=True),
               pd.DataFrame( min_max.fit_transform(df[feature_scale]), columns = feature_scale)],
               axis = 1)
df_new
```

# Feature Selection

```python
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel
```

```python
Y = df_new['revenue']
X = df_new.drop('revenue',axis=1)
```

```python
feature_select = SelectFromModel( Lasso(alpha = 0.005, random_state = 81) )
feature_select.fit(X, Y)
```

```python
feature_select.get_support()
```

```python
sel_features = X.columns[feature_select.get_support()]
sel_features
```

## By using Lasso Regression, we find that none of the variables should be dropped. Later we will use other methods to do the same.

# Now we will apply Multiple Linear Regression

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state=81)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train,Y_train)
```

```python
Y_pred = regressor.predict(X_test)
Y_pred
```

```python
#@title MSE
from sklearn.metrics import mean_squared_error
mse_mul_reg = mean_squared_error(Y_test, Y_pred)
mse_mul_reg
```

```python
errors = abs(Y_pred - Y_test)
print("Mean Absolute Error = ", round(np.mean(errors),2), "degrees")
mape = 100 * errors/Y_test
accuracy = 100 - np.mean(mape)
print("Accuracy = ", round(accuracy,2), "%")
```

# Now we will apply Polynomial Regression

```python
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree = 1)

X_poly_train = poly.fit_transform(X_train)

regressor2 = LinearRegression()

regressor2.fit(X_poly_train, Y_train)
```

```python
X_poly_test = poly.fit_transform(X_test)
```

```python
Y_pred = regressor2.predict(X_poly_test)
Y_pred
```

```python
#@title MSE
mse_poly_reg = mean_squared_error(Y_test, Y_pred)
mse_poly_reg
```

```python
errors = abs(Y_pred - Y_test)
print("Mean Absolute Error = ", round(np.mean(errors),2), "degrees")
mape = 100 * errors/Y_test
accuracy = 100 - np.mean(mape)
print("Accuracy = ", round(accuracy,2), "%")
```

# Decision tree Regression

```python
Y_pred = regressor3.predict(X_test)
```

# Random Forest

```python
features = X.copy()
#features
```

```python
labels = Y.copy()
#labels
```

```python
features_list = list(features.columns)
features = np.array(features)
labels = np.array(labels)
```

```python
train_features, test_features, train_labels, test_labels = train_test_split(features, labels, test_size = 0.4, random_state = 81)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(n_estimators = 1500, random_state = 81)
rf.fit(train_features, train_labels)
```

```python
predictions = rf.predict(test_features)
errors = abs(predictions - test_labels)
print("Mean Absolute Error = ", round(np.mean(errors),2), "degrees")
```

```python
mape = 100 * errors/test_labels
accuracy = 100 - np.mean(mape)
print("Accuracy = ", round(accuracy,2), "%")
```

# Checking for important features

```python
importances = list(rf.feature_importances_)
feature_importances = [(feature, round(importance,2)) for feature,importance in zip(features_list, importances)]
feature_importances = sorted( feature_importances, key = lambda x: x[1], reverse = True )
[print("Variable: {:20} Importance: {}".format(*pair)) for pair in feature_importances]
```

# We see that variables "P29" and "Year" affect the target variable "Revenue" the most.

```python
rf_most_imp = RandomForestRegressor(n_estimators=1500, random_state=81)

imp_indices = [features_list.index('P29'), features_list.index('Year')]
train_imp = train_features[:, imp_indices]
test_imp = test_features[:, imp_indices]

train_imp
rf_most_imp.fit(train_imp, train_labels)
predictions = rf_most_imp.predict(test_imp)
errors = abs(predictions - test_labels)

mape = 100 * (errors/test_labels)
acc = 100 - np.mean(mape)
print("Mean Absolute Error: ",round(np.mean(errors),2))
print("Accuracy: ", round(acc,2),"%")
```

```python
import matplotlib.pyplot as plt

x_values = list(range(len(importances)))
plt.bar(x_values, importances, orientation = 'vertical')

plt.xticks(x_values, features_list, rotation='vertical')

plt.ylabel('Importance'); plt.xlabel('Variable'); plt.title('Variable Importances');
```
