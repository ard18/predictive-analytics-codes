```python
import numpy as np
import pandas as pd
```

```python
from sklearn.datasets import load_breast_cancer
df = pd.DataFrame(load_breast_cancer()['data'],
    columns=load_breast_cancer()['feature_names'])
df['y'] = load_breast_cancer()['target']
```

# Preprocessing

```python
X = df.drop('y', axis=1)
X
```

```python
Y = df['y']
Y
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

```python
minmax = MinMaxScaler()
X_minmax = pd.DataFrame(minmax.fit_transform(X), columns = X.columns)
X_minmax
```

```python
stan = StandardScaler()
X_stan = pd.DataFrame( stan.fit_transform(X), columns = X.columns )
X_stan
```

# Feature Selection

```python
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel
```

```python
feature_select = SelectFromModel( Lasso(alpha = 0.005, random_state = 81) )
feature_select.fit(X, Y)
```

```python
feature_select.get_support()
```

```python
sel_features = X.columns[feature_select.get_support()]
sel_features
```

### We see that
       'mean radius', 'mean texture', 'mean perimeter', 'mean area',
       'perimeter error', 'area error', 'worst radius', 'worst texture',
       'worst perimeter', 'worst area', 'worst concavity'

are the most significant features using Lasso Regression.

```python
X_imp = X[sel_features]
X_imp
```

# Now we start with Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier # Import Decision Tree Classifier
from sklearn.model_selection import train_test_split, GridSearchCV # Import train_test_split function
from sklearn import metrics #Import scikit-learn metrics module for accuracy calculation
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X_imp, Y, test_size = 0.2, random_state = 81)
```

```python
clf = DecisionTreeClassifier()

clf = clf.fit(X_train, Y_train)

Y_pred = clf.predict(X_test)
```

```python
print("Accuracy = ", metrics.accuracy_score(Y_test, Y_pred))
```

```python
#@title Tuning HyperParameters
clf2 = DecisionTreeClassifier()

cv_grid = {'criterion': ["gini","entopy"], 'max_depth': [10, 20, 50, 100], 'min_samples_leaf':[10, 20, 50]}

grid_search = GridSearchCV(estimator=clf2, param_grid=cv_grid, cv=5, scoring='accuracy')  # Adjust scoring metric as needed
grid_search.fit(X_train, Y_train)

grid_search.fit(X_train, Y_train)

best_params = grid_search.best_params_
print("Best hyperparameters:", best_params)

best_model = DecisionTreeClassifier(**best_params)
best_model.fit(X_train, Y_train)

Y_pred = best_model.predict(X_test)
```

```python
print("Accuracy = ", metrics.accuracy_score(Y_test, Y_pred))
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(random_state = 999, warm_start=True)

rf.fit(X_train,Y_train)

Y_pred = rf.predict(X_test)

print("Accuracy = ", metrics.accuracy_score(Y_test, Y_pred))
```

# Gradient Boosting

```python
from sklearn.ensemble import GradientBoostingClassifier

seed = 999
```

```python
gbm = GradientBoostingClassifier(learning_rate=0.2, random_state=seed)

gbm.fit(X_train, Y_train)

Y_pred = gbm.predict(X_test)

print("Accuracy = ", metrics.accuracy_score(Y_test, Y_pred))
```

# KNN Classifier

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors = 2)

knn.fit(X_train, Y_train)

Y_pred = knn.predict(X_test)

print("Accuracy = ", metrics.accuracy_score(Y_test, Y_pred))
```

# Thus, we see that Gradient Boosting does a better job at classification
